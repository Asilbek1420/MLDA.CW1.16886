```python
feature_cols = df.drop('Result', axis=1).columns

df['suspicious_count'] = (df[feature_cols] == -1).sum(axis=1)
df['legit_count'] = (df[feature_cols] == 1).sum(axis=1)
```

```python
inconsistent_legit = df[
    (df['Result'] == 1) &
    (df['suspicious_count'] > df['suspicious_count'].mean() + 2 * df['suspicious_count'].std())
]
```

```python
inconsistent_phishing = df[
    (df['Result'] == -1) &
    (df['legit_count'] > df['legit_count'].mean() + 2 * df['legit_count'].std())
]
print("Inconsistent legitimate-labeled samples:", len(inconsistent_legit))
print("Inconsistent phishing-labeled samples:", len(inconsistent_phishing))
print("\nExamples of inconsistent legitimate labels:\n")
print(inconsistent_legit.head())
print("\nExamples of inconsistent phishing labels:\n")
print(inconsistent_phishing.head())
```

```python
# Importing all relevant libraries for initializing and training the models
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import classification_report, confusion_matrix

X = df.drop(columns=['Result'])
y = df['Result']

# Encodig values to binary to making labeling easier
def encode_binary(label):
    if label == 1:
        return 0          # legitimate
    else:
        return 1          # phishing/suspicious

y_encoded = y.apply(encode_binary)

print(y_encoded[:10])

# Listing all numeric features
numeric_features = X.select_dtypes(include=['int64','float64']).columns.tolist()

# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, stratify=y_encoded, random_state=42)
print("Shape of X and y after splitting int train and test sets")
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Model initialized: RandomForestClassifier
model1 = RandomForestClassifier(n_estimators=200, random_state=42, class_weight='balanced', n_jobs=-1)
#Train the model
model1.fit(X_train, y_train)
```

```python
# XGBoost
model2 = XGBClassifier(n_estimators=200, use_label_encoder=False, eval_metric='logloss',
                                    random_state=42, n_jobs=-1)
model2.fit(X_train, y_train)
```

```python
# CatBoost
model3 = CatBoostClassifier(iterations=500, learning_rate=0.05, depth=6,
                                            random_seed=42, verbose=0)
model3.fit(X_train, y_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model4 = GradientBoostingClassifier()
model4.fit(X_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model5 = KNeighborsClassifier(n_neighbors=7)

model5.fit(X_train, y_train)
```

```python
from sklearn.svm import SVC

model6 = SVC(kernel='linear', probability=True, class_weight='balanced')
model6.fit(X_train, y_train)
```

```python
# Saving models into a file to utilize for StreamLit 
import pickle
pickle.dump(model1, open("models/randomforest.pkl", "wb"))
pickle.dump(model2, open("models/XGBosst.pkl", "wb"))
pickle.dump(model3, open("models/CatBoost.pkl", "wb"))
pickle.dump(model4, open("models/CatBoost.pkl", "wb"))
```

```python
# Creating function where all relevant evaluation metrics are accumulated 
def evaluate_cl_mod(clf, X_test, y_test):
    y_pred = clf.predict(X_test)
    clf_report = classification_report(y_test, y_pred, output_dict=True)
    clf_report = pd.DataFrame(clf_report)
    display(clf_report.T)
    sns.heatmap(confusion_matrix(y_test, y_pred), cbar=False, annot=True)
    plt.show()
```

```python
# Putting models into a list and looping over to evaluate each:
models = [model1, model2, model3, model4, model5, model6]
print("Evaluating all models:")

for m in models:
    print(f"Evaluating {m}:")
    evaluate_cl_mod(m, X_test, y_test)
```
