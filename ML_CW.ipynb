```python
num_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()
num_cols = [c for c in num_cols if c != 'Result'] 

# Plot boxplots
import math
n = len(num_cols)
cols = 3
rows = math.ceil(n/cols)
plt.figure(figsize=(cols*5, rows*4))
for i, col in enumerate(num_cols, 1):
    plt.subplot(rows, cols, i)
    sns.boxplot(y=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
outliers = {}
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    mask = (df[col] < lower) | (df[col] > upper)
    outliers[col] = mask.sum()
sorted(outliers.items(), key=lambda x: x[1], reverse=True)[:10]
```

```python
from scipy.stats.mstats import winsorize
df_w = df.copy()
for col in num_cols:
    # clip to 1st and 99th percentiles
    low, high = df[col].quantile([0.01, 0.99]).values
    df_w[col] = df[col].clip(low, high)
```

```python
corr = df[num_cols + ['Result']].corr()
plt.figure(figsize=(12,10))
sns.heatmap(corr, annot=False, cmap='coolwarm')
plt.title('Correlation heatmap')
plt.show()

# View high correlations
threshold = 0.9
high_corr = []
for i in range(len(num_cols)):
    for j in range(i+1, len(num_cols)):
        if abs(corr.iloc[i,j]) > threshold:
            high_corr.append((num_cols[i], num_cols[j], corr.iloc[i,j]))
high_corr
```
